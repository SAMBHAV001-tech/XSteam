## Conclusions

### Task 1: Sentiment Classification Performance
The **Logistic Regression** model performed the best for this task. It achieved the highest Accuracy, F1-Score, and ROC-AUC metrics overall. Linear models like Logistic Regression often perform exceptionally well on high-dimensional sparse continuous representations like TF-IDF, while being less prone to overfitting and much faster to train and infer compared to tree-based models like XGBoost.

### Task 2: Helpfulness Prediction
The features that mattered most for helpfulness prediction included:
1. **sentiment_score**: Reviews with strong sentiments (very positive or very negative scores inferred from LR model) tend to be voted as more helpful by readers because they convey clearer judgments.
2. **review_length / word_count**: Longer, more detailed reviews obviously provide much more context, nuance, and evaluation to the community, making them inherently more useful than short one-liners.
3. **playtime_hours**: Users who have played the game extensively often provide more credible, grounded, and insightful feedback, causing other users to upvote them.

### 3 Key EDA Insights Discovered
1. **Class Imbalance in Sentiment**: There is a heavy skew towards positive reviews (recommended=True). Therefore, making balanced class weights during modeling is mandatory.
2. **Right-Skewed Helpfulness Distribution**: Most reviews receive 0 or very few 'helpful' votes, while a tiny fraction goes viral with thousands of votes. Applying a log scale during analytics is important to visualize this heavy power-law tail.
3. **Review Length Distribution**: Negative reviews tend to occasionally be longer or more intensely detailed as users construct essays to justify their dissatisfaction, although general standard reviews peak pretty uniformly at a moderate character limit.
