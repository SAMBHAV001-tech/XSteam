```python
import pandas as pd
import numpy as np
import os
import sys

# Add src to path
sys.path.append(os.path.abspath(os.path.join('..', 'src')))
from preprocess import clean_text
from features import engineer_features
```

```python
# Load raw CSV
data_path = os.path.join('..', 'data', 'raw', 'steam_reviews.csv')
print(f"Loading data from {data_path}")
df = pd.read_csv(data_path)
```

```python
# Sample 100,000 rows for speed
if len(df) > 100000:
    print("Sampling 100,000 rows randomly for speed processing")
    df = df.sample(100000, random_state=42)
```

```python
# Apply clean_text() to review column
print("Cleaning text...")
# Handling missing data to avoid errors
df['review_text'] = df['review_text'].fillna('')
df['cleaned_review'] = df['review_text'].apply(clean_text)

# Drop rows where cleaned_review is null or empty
df = df.dropna(subset=['cleaned_review'])
df = df[df['cleaned_review'].str.strip() != '']
```

```python
# Apply engineer_features()
print("Engineering features...")
df = engineer_features(df)
```

```python
# Save final DataFrame to data/processed/steam_cleaned.csv
out_path = os.path.join('..', 'data', 'processed', 'steam_cleaned.csv')
os.makedirs(os.path.dirname(out_path), exist_ok=True)
df.to_csv(out_path, index=False)
print(f"Saved cleaned data to {out_path}")
```
